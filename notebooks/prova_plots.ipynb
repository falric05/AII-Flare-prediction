```python
from Libs.load_data import DataLoader
from Libs import flares_plot as fplt
from Libs.threshold import get_labels_physics
```

```python
# initialize data loader
data_loader = DataLoader()
# get the grid
grid = data_loader.get_grid()
```

```python
labels = get_labels_physics(grid, data_loader.get_params())
```

```python
# extract the standard values
run = 5
bindeces = data_loader.get_standard_indeces()
X_best = data_loader.get_standard_values(run=run)
labels_best = labels[run, bindeces[0], bindeces[1], bindeces[2], bindeces[3], :]
# plot one single light curve
fplt.plot_serie(X_best, labels=labels_best)
```

```python
# plot grid varying one single parameter
fplt.plot_grid_datapoints(grid, data_loader.get_params(), run=6, variation='theta')
```
